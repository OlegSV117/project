```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
t=[]
x = []
y = []
a=10
b=3
for i in range(101):
    t_ = i*2*3,14/100
    x.append(a*np.cos(t_))
    y.append(b*np.sin(t_)) 
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
```
